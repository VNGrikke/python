```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu
x = np.linspace(0, 10, 4)    
y = x**2 + 2*x

# Tạo Figure và Axes
fig, ax = plt.subplots(figsize=(7, 4))

# Vẽ đường
ax.plot(x, y, linewidth=2)

# Thiết lập tiêu đề và nhãn
ax.set_title("Biểu đồ đầu tay của Vuong")
ax.set_xlabel("Trục x")
ax.set_ylabel("Trục y")

# Bật lưới dạng nét đứt
ax.grid(True, linestyle='--')

plt.tight_layout()
plt.show()
```

```python
# 3
import numpy as np
import matplotlib.pyplot as plt

# Sinh dữ liệu từ phân phối chuẩn
data = np.random.normal(loc=0, scale=1, size=1000)

# Tạo Figure và Axes
fig, ax = plt.subplots(figsize=(6, 4))

# Vẽ histogram
ax.hist(data, bins=30, edgecolor='black')

# Tiêu đề và nhãn
ax.set_title("Histogram phân phối chuẩn")
ax.set_xlabel("Giá trị")
ax.set_ylabel("Tần suất")

ax.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu
x = np.linspace(0, 10, 400)
y1 = x
y2 = x**2
y3 = x**3

fig, ax = plt.subplots(figsize=(7, 4))

# Vẽ 3 đường với màu khác nhau
ax.plot(x, y1, label="y = x")
ax.plot(x, y2, label="y = x²")
ax.plot(x, y3, label="y = x³")

# Thiết lập tiêu đề và nhãn
ax.set_title("So sánh các hàm bậc 1, 2 và 3")
ax.set_xlabel("x")
ax.set_ylabel("Giá trị hàm số")

# Lưới & legend
ax.grid(True, linestyle="--", alpha=0.4)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu cho lineplot
x = np.linspace(0, 10, 300)
y1 = x
y2 = x**2

# Dữ liệu cho scatter
xs = np.random.uniform(0, 50, 100)
ys = np.random.uniform(0, 50, 100)

# Tạo figure 2x2
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Subplot (0,0) – lineplot y=x và y=x²
axs[0, 0].plot(x, y1, label="y = x")
axs[0, 0].plot(x, y2, label="y = x²")
axs[0, 0].set_title("Lineplot: y = x và y = x²")
axs[0, 0].legend()
axs[0, 0].grid(True, linestyle='--', alpha=0.4)

# Subplot (0,1) – scatter 1
axs[0, 1].scatter(xs, ys, alpha=0.7)
axs[0, 1].set_title("Scatterplot ngẫu nhiên 1")
axs[0, 1].grid(True, linestyle='--', alpha=0.4)

# Subplot (1,0) – scatter 2 (dữ liệu khác để không trùng)
xs2 = np.random.uniform(0, 50, 100)
ys2 = np.random.uniform(0, 50, 100)

axs[1, 0].scatter(xs2, ys2, alpha=0.7)
axs[1, 0].set_title("Scatterplot ngẫu nhiên 2")
axs[1, 0].grid(True, linestyle='--', alpha=0.4)

# Subplot (1,1) – lineplot lại (tuỳ chỉnh)
axs[1, 1].plot(x, y1, color='green')
axs[1, 1].set_title("Lineplot: y = x (phiên bản khác)")
axs[1, 1].grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(7, 4))

# Vẽ đường sin(x)
ax.plot(x, y, label="y = sin(x)")

# Điểm cần annotate
x0 = np.pi / 2
y0 = np.sin(x0)

# Chú thích
ax.annotate(
    "Cực đại tại x = π/2", 
    xy=(x0, y0),                     # điểm cần đánh dấu
    xytext=(x0 + 0.5, y0 + 0.3),     # vị trí text
    arrowprops=dict(arrowstyle="->", lw=1.5)
)

# Tiêu đề, nhãn
ax.set_title("Đường y = sin(x) và chú thích cực đại")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")

ax.grid(True, linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Tạo ma trận ngẫu nhiên 10x10
data = np.random.uniform(0, 100, (10, 10))

fig, ax = plt.subplots(figsize=(7, 6))

# Vẽ heatmap
sns.heatmap(
    data,
    ax=ax,
    cmap="viridis",     # bảng màu trực quan tốt
    annot=True,         # hiển thị giá trị trong từng ô
    fmt=".1f",          # hiển thị 1 chữ số thập phân
    linewidths=0.5,     
    cbar=True
)

# Tiêu đề & nhãn trục
ax.set_title("Heatmap từ ma trận 10x10 (ngẫu nhiên)")
ax.set_xlabel("Cột")
ax.set_ylabel("Hàng")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Dữ liệu cố định
x = np.linspace(0, 2*np.pi, 400)

fig, ax = plt.subplots(figsize=(7, 4))

# Khởi tạo đường trống
line, = ax.plot([], [], lw=2)

# Giới hạn trục
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.2, 1.2)

# Tiêu đề và nhãn
ax.set_title("Sóng sin động với FuncAnimation")
ax.set_xlabel("x")
ax.set_ylabel("sin(x + t)")

# Hàm khởi tạo
def init():
    line.set_data([], [])
    return line,

# Hàm cập nhật theo frame
def animate(frame):
    y = np.sin(x + 0.1 * frame)
    line.set_data(x, y)
    return line,

# Tạo animation
anim = FuncAnimation(fig, animate, init_func=init,
                     frames=200, interval=30, blit=True)

plt.show()
```
