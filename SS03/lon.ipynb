```python
# bt1
def sub (a, b ):
    return a-b

def mul (a, b ):
    return a*b

a = int (input ("nhap a: "))
b = int (input ("nhap b: "))

print ("a - b =", sub (a, b))
print ("a * b =", mul (a, b))
```

```python
# bt2
def isPrime(a):
    if a < 2:
        return False
    for i in range(2, int(a**0.5) + 1):
        if a % i == 0:
            return False
    return True

a = int(input("nhap a: "))
if isPrime(a): print(a, "la so nguyen to")
else: print(a, "khong phai so nguyen to")
```

```python
# bt3
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
n = int(input("nhap n: "))
print("giai thua cua", n, "là", factorial(n))
```

```python
# bt4
def circle_area(r):
    pi = 3.14159
    return pi * r * r

r = float(input("nhap ban kinh hinh tron: "))
print("dien tich hinh tron la:", circle_area(r))
```

```python
# bt5
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
    
year = int(input("Nhap nam: "))
if is_leap_year(year): print(year, "la nam nhuan")
else: print(year, "khong phai nam nhuan")
    
```

```python
# bt6
def sum(n):
    total = 0
    number = list(n)
    for i in number:
        total += int(i)
    return total

n = input("Nhap n: ")
print("Tong cac chu so cua", n, "la:", sum(n))
```

```python
# bt7
def handleReverseTheNumber(number):
    reNumber = reversed(number)
    return ''.join(reNumber)

try:
    number = int(input("Nhap so nguyen duong: "))
    if number <= 0:
        print("Vui long nhap so nguyen duong!")
    else:
        print("So dao nguoc la:", handleReverseTheNumber(str(number)))

except ValueError:
    print("Vui long nhap mot so nguyen hop le!")
```

```python
# bt8
def perfectNumber(n):
    sum_divisors = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            sum_divisors += i
            if i != n // i:
                sum_divisors += n // i
    return sum_divisors == n

n = int(input("Nhap n: "))
if perfectNumber(n):
    print(n, "la so hoan hao")
else:
    print(n, "khong phai so hoan hao")
```
