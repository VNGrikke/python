```python
products = [
    ("A001", "Laptop", 1500),
    ("A002", "Smartphone", 800),
    ("A003", "Tablet", 600),
    ("A004", "Monitor", 300),
    ("A005", "Keyboard", 100)
]

def displayMenu():
    while True:
        print("----------------------------------------")
        print("                 Menu")
        print("1. Thêm sản phẩm mới.")
        print("2. Hiển thị danh sách sản phẩm.")
        print("3. Tìm kiếm sản phẩm theo mã sản phẩm.")
        print("4. Cập nhật thông tin sản phẩm.")
        print("5. Xóa sản phẩm theo mã sản phẩm.")
        print("0. Thoát chương trình.")
        
        choice = int(input("Chọn chức năng: "))
        
        if choice == 0:
            break
        
        elif choice == 1:
            add_product(products)
        
        elif choice == 2:
            display_products(products)
        
        elif choice == 3:
            code = input("Nhập mã sản phẩm cần tìm: ")
            product = search_product(products, code)
            if product:
                print(f"Tìm thấy sản phẩm: Mã: {product[0]}, Tên: {product[1]}, Giá: {product[2]}")
            else:
                print("Không tìm thấy sản phẩm.")
        
        elif choice == 4:
            code = input("Nhập mã sản phẩm cần cập nhật: ")
            update_product(products, code)
        
        elif choice == 5:
            code = input("Nhập mã sản phẩm cần xóa: ")
            delete_product(products, code)
        
        else:
            print("Chức năng không hợp lệ.")


def add_product(products):
    code = input("Nhập mã sản phẩm: ")
    while duplicate_code(products, code):
        print("Mã sản phẩm đã tồn tại. Vui lòng nhập mã khác.")
        code = input("Nhập mã sản phẩm: ")
    name = input("Nhập tên sản phẩm: ")
    price = float(input("Nhập giá sản phẩm: "))
    products.append((code, name, price))
    print("Thêm sản phẩm thành công.")

def duplicate_code(products, code):
    for product in products:
        if product[0] == code:
            return True
    return False

def display_products(products):
    print("Danh sách sản phẩm:")
    for product in products:
        print(f"Mã: {product[0]}, Tên: {product[1]}, Giá: {product[2]}")


def search_product(products, code):
    for product in products:
        if product[0] == code:
            return product
    return None


def update_product(products, code):
    for i in range(len(products)):
        if products[i][0] == code:
            name = input("Nhập tên sản phẩm mới: ")
            price = float(input("Nhập giá sản phẩm mới: "))
            products[i] = (code, name, price)
            print("Cập nhật sản phẩm thành công.")
            return
    print("Không tìm thấy mã sản phẩm.")


def delete_product(products, code):
    for i in range(len(products)):
        if products[i][0] == code:
            products.pop(i)
            print("Xóa sản phẩm thành công.")
            return
    print("Không tìm thấy mã sản phẩm.")


displayMenu()
```
